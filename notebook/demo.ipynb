```python
movies = pd.read_csv('../data/movielens/movies.csv', encoding="ISO-8859-1")
movies['item_title'] = movies['item_title'].apply(lambda x: ' '.join(x.split()[:-1]))
movies.head()
```

```python
users = pd.read_csv('../data/movielens/users.csv', encoding="ISO-8859-1")
users['gender'] = (users['gender'] == 'M').astype(int)
users.head()
```

```python
ratings = pd.read_csv('../data/movielens/ratings.csv', encoding="ISO-8859-1")
ratings.head()
```

```python
users['rating_count'] = ratings['userid'].value_counts().sort_index().values
movies['rating_count'] = ratings['itemid'].value_counts().sort_index().values
```

```python
movies.head()
```

```python
genres = list(movies.columns.drop(['itemid', 'item_title']))
genres_dict = dict((k,v) for v,k in enumerate(genres))
genres_dict['Family'] = 19
genres_dict['Short'] = 20
genres_dict
```

```python
imdb_basics = pd.read_csv('../data/imdb/title.basics.tsv', encoding="ISO-8859-1", delimiter='\t')
imdb_basics = imdb_basics.drop_duplicates('primaryTitle')
imdb_basics = imdb_basics[imdb_basics['titleType'] == 'movie']
print(imdb_basics.shape)
imdb_basics.head()
```

```python
imdb_basics['genre'] = imdb_basics['genres'].apply(lambda x: genres_dict[x.split(',')[0]] if type(x) == np.str and x.split(',')[0] in genres_dict else -1)
```

```python
imdb_basics['runtime'] = imdb_basics['runtimeMinutes'].apply(lambda x: int(x) if x.isdigit() else np.nan)
```

```python
imdb_b_filter = imdb_basics[imdb_basics['originalTitle'].isin(movies['item_title'])]
imdb_b_filter.info()
imdb_b_filter.head()
```

```python
mdata_left = movies[['itemid', 'item_title', 'rating_count']]
mdata_right = imdb_b_filter[['primaryTitle', 'runtimeMinutes', 'genre', 'runtime']]
```

```python
mdata = pd.merge(mdata_left, mdata_right, left_on='item_title', right_on='primaryTitle')
mdata = mdata[mdata.columns.drop(['item_title', 'primaryTitle', 'runtimeMinutes'])]
mdata = mdata.dropna()
mdata.head()
```

```python
mratings = ratings[ratings['itemid'].isin(mdata['itemid'])]
```

```python
musers = users[users.columns.drop(['occupation', 'zipcode'])]
```
